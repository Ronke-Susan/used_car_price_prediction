### Heatmap showing correlation between dataset features

```python
sns.set(rc={'figure.figsize':(12.7,9.27)})
sns.heatmap(df.corr(),annot=True,fmt='.2f')
plt.show()
```

* There is a strong negative correlation value of -0.75 between car production year and car mileage
* There is a positve correlation value of 0.52 between car production year and car price
* There is a positive correlation value of 0.61  between car price and engine size
* There is a strong negative correlation value of -0.88 between petrol and diesel
* There is a week correlation between car gear types (automatic, semi-auto, manual, etc) and car price

### Price distribution

```python
sns.set_style("whitegrid")
sns.set(rc={'figure.figsize':(12.7,9.27)})
sns.histplot(df['price'], color="red")
plt.show()
```

Price data is skewed to the right ( positve skewness)

### Model vs Price

```python
mode_l=pd.DataFrame(df.groupby('model')['price'].mean()).reset_index(inplace= False)
mode_l
```

```python
mode_l.sort_values("price", ascending=False)
```

```python
model_sorted= mode_l.sort_values("price", ascending=False)
fig= plt.figure(figsize=(20,8))
# make bar plot with sns
sns.set_style("whitegrid")
sns_plot = sns.barplot(x = 'model', y = 'price', data = model_sorted, color='blue')
plt.xlabel("model")
plt.ylabel("price")
plt.xticks(rotation=90)
plt.title("Average price for each car model", size=18)
plt.show()
#fig.savefig('Car model prices.jpg')
```

The top four most expensive car models are G class, R8, X7, and 8 series. They cost above 60,000 dollars

### Year vs Price

```python
year_price = pd.DataFrame(df.groupby('year')['price'].mean()).sort_values('price',ascending=False).reset_index(inplace= False)
year_price
```

```python
fig= plt.figure(figsize=(15, 7))
plt.plot(df.groupby('year')['price'].mean(), marker='d', color='black')
plt.title('Average car price by Year')
plt.xlabel('Year')
plt.ylabel('price')
plt.grid()
plt.show()
```

```python
fig= plt.figure(figsize=(20, 9))
plt.bar('year','price', data= year_price)
plt.title('Average car price by year', fontsize=20)
plt.xlabel('year',fontsize=18)
plt.ylabel('price',fontsize=16)
plt.show()
```

car produced in recent years cost more than cars produced in former years with the exception of vintage cars. The more recent the production year, the more expensive the price of the car

### Mileage vs Price

```python
mile_age= pd.DataFrame(df.groupby('model')['mileage'].mean()).sort_values('mileage',ascending=False).reset_index(inplace= False)
mile_age
```

```python
sns.set_style("whitegrid", {'axes.grid': False})
fig= plt.figure(figsize=(20, 9))
sns_plot = sns.barplot(x = 'model', y = 'mileage', data = mile_age, color='magenta')
plt.xlabel("model")
plt.ylabel("mileage")
plt.xticks(rotation=90)
plt.title("average mileage for each car model", size=18)
plt.show()
```
